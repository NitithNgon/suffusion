```python
import cv2
import os
import math
def extract_frames(video_path, output_folder, interval=1):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video file")
        return

    frame_count = 0
    fps = cap.get(cv2.CAP_PROP_FPS)
    print ("Actual frame rate: ",fps)
    fps = math.ceil(fps)
    print ("Calculate fps:", fps)
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame_count += 1
        #print ("current frame", frame_count)

        # Calculate time in seconds
        current_time = frame_count / fps
        #print (current_time)
        # Check if the current time is a multiple of the interval
        if frame_count % fps == 0:
            # Save the frame
            output_path = os.path.join(output_folder, f"frame_{frame_count}.jpg")
            cv2.imwrite(output_path, frame)

    # Release video capture object
    cap.release()

    print(f"Frames extracted at {interval} second interval and saved to {output_folder}")

# Example usage
video_path = r"C:\Users\User\Desktop\Thesis\suffusion\python\2024_03_24_16_47_IMG_9262.MOV"
output_folder = r"C:\Users\User\Desktop\Thesis\suffusion\python\asset\IMG_SQ"
interval = 1  # Extract frames every 1 second

extract_frames(video_path, output_folder, interval)
```
